# __Generate Training Set__
### In the below code i am generating a random set of data which will be tested by using LinearRegression and Predictions.

> * ## Using the random import to create random Math values to be generated for the data - all data are integers and contain no fractional values.

```python
from random import randint
# Setting limits
TRAIN_SET_LIMIT = 1000
TRAIN_SET_COUNT = 100
# Initialise as lists
TRAIN_INPUT = list()
TRAIN_OUTPUT = list()

for i in range(TRAIN_SET_COUNT):
    a = randint(0, TRAIN_SET_LIMIT)
    b = randint(0, TRAIN_SET_LIMIT)
    c = randint(0, TRAIN_SET_LIMIT)
    
    op = a + (2*b) + (3*c)
    # Append to list
    TRAIN_INPUT.append([a, b, c])
    TRAIN_OUTPUT.append(op)
    print(TRAIN_OUTPUT)
    # for
```

# Linear Regression Model
> ## CREATE a Model, TRAIN it and then USE it

* ### First we create the Linear Regression Model and then we pass it the training data that we generated up above.

```python
from sklearn.linear_model import LinearRegression

predictor = LinearRegression(n_jobs=-1)
predictor.fit(X=TRAIN_INPUT, y=TRAIN_OUTPUT)
print("Model Created and data has been passed.")
```

# Test Data
> ## x = [[10,20,30]] so the equation should work out to be (10+(20*2)+(30*3)) = 140!

```python
X_TEST = [[10, 20, 30]]
outcome = predictor.predict(X=X_TEST)
coefficients = predictor.coef_

print("Outcome : {}\nCoefficients : {}".format(outcome, coefficients))
print("You can change the above values inside the X_TEST list to test the equation for how accurate it is.")
```
