```python
import sklearn
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
from sklearn import linear_model, preprocessing
```

```python
# read in the data file
data = pd.read_csv("car.data")
```

```python
# display head-top 5 lines
print(data.head())
```

```python
# create object
le = preprocessing.LabelEncoder()

# returning a numpy array
buying = le.fit_transform(list(data["buying"]))
maint = le.fit_transform(list(data["maint"]))
door = le.fit_transform(list(data["door"]))
persons = le.fit_transform(list(data["persons"]))
lug_boot = le.fit_transform(list(data["lug_boot"]))
safety = le.fit_transform(list(data["safety"]))
classobj = le.fit_transform(list(data["class"]))
```

```python
predict = "class"
```

```python
X = list(zip(buying, maint, door, persons, lug_boot, safety)) # convert to list
y = list(classobj) #  convert to list
```

```python
# train and test assign
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(X,y,test_size=0.1)
```

```python
print(x_train, y_test)
```

```python
model = KNeighborsClassifier(n_neighbors=9)

# train the model
model.fit(x_train, y_train)
```

```python
# accuracy test
acc = model.score(x_test, y_test)
print(acc)


# predict the data
predicted = model.predict(x_test)
```

```python
names = ["unacc", "acc", "good", "vgood"]

for x in range(len(predicted)):
    print("Predicted : ",names[predicted[x]], "Data: " , x_test[x], "Actual : ", names[y_test[x]])
    # pass 2d array - only want 1 data so nest array inside array
    n = model.kneighbors([x_test[x]], 9, True)
    print("N : ",n)
```
